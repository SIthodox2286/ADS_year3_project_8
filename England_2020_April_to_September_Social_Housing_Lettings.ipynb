```python
import pandas as pd
```

```python
def createDF(sheet, skipRows, skipFooter, header):
    if header:
        df = pd.read_excel (r"./Data/England_2020_April_to_September_Social_Housing_Lettings.xlsx", sheet_name=sheet,  skiprows=skipRows, skipfooter = skipFooter,  header=header)
    else:
        df = pd.read_excel (r"./Data/England_2020_April_to_September_Social_Housing_Lettings.xlsx", sheet_name=sheet,  skiprows=skipRows, skipfooter = skipFooter)
    df.dropna(how='all', inplace=True)
    df.dropna(how='all', axis=1, inplace=True)
    
    if header:
        head = []
        for row in range(len(header)):
            a = df.columns.get_level_values(row).to_series()
            b = a.mask(a.str.startswith('Unnamed')).ffill().fillna('')
            head.append(b)
        df.columns = head
    return df
```
